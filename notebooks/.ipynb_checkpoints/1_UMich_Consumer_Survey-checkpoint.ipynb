```python
import pandas as pd
import os
import sys
current_dir = os.getcwd()
utils_dir = os.path.join(current_dir, '..', 'utils')
sys.path.append(utils_dir)
from chartspecs import *
```

### University of Michigan Survey of Consumers

```python
### University of Michigan Consumer Sentiment
data_url = 'http://www.sca.isr.umich.edu/files/tbmics.csv'
df = pd.read_csv(data_url)
df['date'] = df['Month'] + ' ' + df['YYYY'].astype(str)
df['date'] = pd.to_datetime(df['date'], format='%B %Y')
df.set_index('date', inplace=True)
data = df.loc[:, ['ICS_ALL']]
data.columns = ['U.Mich: Consumer sentiment']
file_path = os.path.join(data_dir, 'umich_cons_senti.csv')
data.to_csv(file_path, index_label='date', header=True)
data
```

```python
d = {}
for i in [-1, -2, -13]:
    d[i] = {'date': dtxt(data.index[i])['mon1'], 'val': data.iloc[i].iloc[0]}

pcv = data.loc['2019-03-01':'2020-02-01'].mean().iloc[0]
ch_val = -(1 - (d[-1]['val'] / pcv)) * 100
ch_txt = value_text(ch_val, style='above_below', threshold=0.5)
```

```python

text = (f'As of {d[-1]["date"]}, the latest value of the consumer sentiment index is '+
        f'{d[-1]["val"]}, following {d[-2]["val"]} in {d[-2]["date"]}, and compared to '+
        f'{d[-13]["val"]} one year prior, in {d[-13]["date"]}. '+
        f'As a pre-COVID baseline, the index average value was {pcv:.1f} '+
        'during the year ending February 2020; the consumer '+
        f'sentiment index is currently {ch_txt} this level.')
file_path = os.path.join(text_dir, 'umichsoc.txt')
write_txt(file_path, text)
print(text)
```

```python
# Plot the data
footnote_params = {
    'xy': (0,-0.05),
    'xycoords':'axes fraction',
    'xytext': (0,-24),
    'textcoords': 'offset points',
    'va': 'top',
}
fig, ax = plt.subplots(figsize=(12,6))
df1 = data.loc['2018':].copy()
for i, item in enumerate(list(df1.columns)):
    df_ = df1[[item]]
    ax.plot(df_.index.values, df_[item], label = item, linewidth=2.5)
ax1 = ax.twinx()
ax1.set_ylim(ax.get_ylim())
start_date = df1.index.min() 
end_date = df1.index.max()  
end_date_plus_12_months = end_date + relativedelta(months=12)
ax.set_xlim(start_date, end_date_plus_12_months)

ax.grid(**grid_kwargs)
fig.text(0.125,0.90,"Index value",fontdict={'fontsize':15,'fontweight':'regular','fontname':'Baskerville'})
footnote_text = "Source: University of Michigan, AllStuffData"
ax.set_title('UMich. Consumer Sentiment', loc='left', fontdict={'fontsize':25,'fontweight':'bold','fontname':'Baskerville'}, pad=30 )
ax.set_xlabel("Date")
#ax.legend(LEGEND_KWARGS,legend_placement_kwargs)
apply_to_axes((ax, ax1), footnote_text=footnote_text)
footnote = ax.annotate(footnote_text, **footnote_params)
footnote.set_fontsize(footnote_fontsize)
footnote.set_fontname(fontname)
footnote.set_fontweight(footnote_fontweight)
file_path = os.path.join(charts_dir, 'umich_consumer_sentiment.jpg')
fig.savefig(file_path, **savefig_params)
```

```python
### Current conditions and expections
data_url = 'http://www.sca.isr.umich.edu/files/tbmiccice.csv'
df = pd.read_csv(data_url)
df['date'] = pd.to_datetime(df['Month'] + ' ' + df['YYYY'].astype(str), format='%B %Y')
df.set_index('date', inplace=True)
selected_columns = ['ICC', 'ICE']
df_selected = df[selected_columns]
df_selected.columns = ['U.Mich: Current Conditions', 'U.Mich: Consumer Expectations']
file_path = os.path.join(data_dir, 'umich_cond_expec.csv')
df_selected.to_csv(file_path, index_label='date', header=True)
df_selected
```

```python
df = df_selected.copy()
df.columns = ['ICC', 'ICE']
ltdt = dtxt(df.index[-1])['mon1']
prdt = dtxt(df.index[-2])['mon1']
ccval = df.ICC.iloc[-1]
ccpr = df.ICC.iloc[-2]
cc19 = df.loc['2019', 'ICC'].mean()
```

```python
ceval = df.ICE.iloc[-1]
cepr = df.ICE.iloc[-2]
ce19 = df.loc['2019', 'ICE'].mean()

text = ('The consumer sentiment index combines views on current '+
        f'and future economic conditions. In {ltdt}, the index tracking '+
        f'views on current economic conditions was {ccval:.1f}, '+
        f'compared to {ccpr:.1f} in {prdt}, and {cc19:.1f} '+
        f'in 2019.'+
        f'In {ltdt}, the index tracking consumer '+
        f'expectations for future economic conditions was {ceval:.1f}, '+
        f'compared to {cepr:.1f} in {prdt}, and {ce19:.1f} '+
        f'in 2019.')
file_path = os.path.join(text_dir, 'umichsoc_ice.txt')
write_txt(file_path, text)
print(text)
```

```python
# Plot the data
footnote_params = {
    'xy': (0,-0.05),
    'xycoords':'axes fraction',
    'xytext': (0,-24),
    'textcoords': 'offset points',
    'va': 'top',
    
}
legend_placement_kwargs = {
    'loc': 'upper center',
    'bbox_to_anchor': (.8,.98),
    'ncol': 1,
}
fig, ax = plt.subplots(figsize=(12,6))
df1 = df_selected.loc['2018':].copy()
for i, item in enumerate(list(df1.columns)):
    df_ = df1[[item]]
    ax.plot(df_.index.values, df_[item], label = item, linewidth=2.5)
ax1 = ax.twinx()
ax1.set_ylim(ax.get_ylim())
start_date = df1.index.min() 
end_date = df1.index.max()  
end_date_plus_12_months = end_date + relativedelta(months=12)
ax.set_xlim(start_date, end_date_plus_12_months)

ax.grid(**grid_kwargs)
fig.text(0.125,0.90,"Index value",fontdict={'fontsize':15,'fontweight':'regular','fontname':'Baskerville'})
footnote_text = "Source: University of Michigan, AllStuffData"
ax.set_title('UMich. Current conditions & Consumer expectations', loc='left', fontdict={'fontsize':25,'fontweight':'bold','fontname':'Baskerville'}, pad=30 )
ax.set_xlabel("Date")
ax.legend(**LEGEND_KWARGS,**legend_placement_kwargs)
apply_to_axes((ax, ax1), footnote_text=footnote_text)
footnote = ax.annotate(footnote_text, **footnote_params)
footnote.set_fontsize(footnote_fontsize)
footnote.set_fontname(fontname)
footnote.set_fontweight(footnote_fontweight)
file_path = os.path.join(charts_dir, 'umich_current_conditions.jpg')
fig.savefig(file_path, **savefig_params)
```

```python
### Inflation Expecations (One and Five Years Ahead)
# Historical data
historical_url = 'http://www.sca.isr.umich.edu/files/tbmpx1px5.csv'
historical_df = pd.read_csv(historical_url)
historical_df.index = pd.to_datetime(historical_df['Month'] + ' ' + historical_df['YYYY'].astype(str), format='%B %Y')
historical_df.drop(['Month', 'YYYY'], axis=1, inplace=True)
historical_df
```

```python
# Latest data
latest_url = 'http://www.sca.isr.umich.edu/files/tbcpx1px5.csv'
latest_df = pd.read_csv(latest_url, skiprows=3).dropna(axis=1, how='all').dropna()
latest_df.columns = ['Month', 'YYYY', 'PX_MD', 'PX5_MD']
latest_df['Month'] = latest_df['Month'].str.extract(r'([a-zA-Z]+)')
latest_df.index = pd.to_datetime(latest_df['Month'] + ' ' + latest_df['YYYY'].astype(int).astype(str), format='%B %Y')
latest_df.drop(['Month', 'YYYY'], axis=1, inplace=True)
latest_df
```

```python
#There might be some rows which might be missing in historical data so we are doing this.
```

```python
# Combine historical and latest data
if latest_df.index[-1] not in historical_df.index:
    historical_df = pd.concat([historical_df, latest_df.iloc[-1:].copy()])

historical_df.columns = ['U.Mich: Inflation expectations (1 year ahead)','U.Mich: Inflation expectations (5 years ahead)']
file_path = os.path.join(data_dir, 'umich_infl_expec.csv')
historical_df.loc[:].to_csv(file_path, index_label='date')
```

```python
historical_df.tail(5)
```

```python
dfm = historical_df.copy()
dfm.columns = ['PX_MD','PX5_MD']
data = dfm['PX5_MD'].loc['2017':].copy()
ltdt = dtxt(data.index[-1])['mon1']
prdt = dtxt(data.index[-13])['mon1']
p5val = data.iloc[-61]
lval = data.iloc[-1]
pval = data.iloc[-13]
inf_act = pd.read_csv(os.path.join(data_dir, 'cpi.csv'))['ALL'].iloc[-60:].mean()
```

```python
text = (f'As of {ltdt}, surveyed consumers expect inflation to average '+
        f'{lval} percent over the next five years, '+
        f'compared to an expected rate of {pval} percent in {prdt}. '+
        f'Consumers had expected inflation to average {p5val} percent over '+
        f'the past five years, while actual inflation over the period '+
        f'was {inf_act:.1f} percent.')
write_txt(os.path.join(text_dir, 'inf_exp_cons.txt'), text)
print(text)
```

```python
yrval = dfm['PX_MD'].iloc[-1]
text = (f'Respondents expect consumer prices to increase {yrval} '+
        f'percent over the year starting {ltdt}.')
write_txt(os.path.join(text_dir, 'inf_exp_surv_st.txt'), text)
print(text)
```

```python
# Plot the data
footnote_params = {
    'xy': (0,-0.05),
    'xycoords':'axes fraction',
    'xytext': (0,-24),
    'textcoords': 'offset points',
    'va': 'top',
    
}
legend_placement_kwargs = {
    'loc': 'upper center',
    'bbox_to_anchor': (.25,.98),
    'ncol': 1,
}
fig, ax = plt.subplots(figsize=(12,6))
df1 = historical_df.loc['2018':].copy()
for i, item in enumerate(list(df1.columns)):
    df_ = df1[[item]]
    ax.plot(df_.index.values, df_[item], label = item, linewidth=2.5)
ax1 = ax.twinx()
ax1.set_ylim(ax.get_ylim())
start_date = df1.index.min() 
end_date = df1.index.max()  
end_date_plus_12_months = end_date + relativedelta(months=12)
ax.set_xlim(start_date, end_date_plus_12_months)

ax.grid(**grid_kwargs)
fig.text(0.125,0.90,"Index value",fontdict={'fontsize':15,'fontweight':'regular','fontname':'Baskerville'})
footnote_text = "Source: University of Michigan, AllStuffData"
ax.set_title('UMich. Inflation expectations', loc='left', fontdict={'fontsize':25,'fontweight':'bold','fontname':'Baskerville'}, pad=30 )
ax.set_xlabel("Date")
ax.legend(**LEGEND_KWARGS,**legend_placement_kwargs)
apply_to_axes((ax, ax1), footnote_text=footnote_text)
footnote = ax.annotate(footnote_text, **footnote_params)
footnote.set_fontsize(footnote_fontsize)
footnote.set_fontname(fontname)
footnote.set_fontweight(footnote_fontweight)
file_path = os.path.join(charts_dir, 'umich_inflation_expectations.jpg')
fig.savefig(file_path, **savefig_params)
```
